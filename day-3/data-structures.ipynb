### 🏋️ Exercise — Lists

You have a list of exam scores. Write code to:
1. Sort the scores in **descending** order and print them
2. Print the **top 3** scores
3. Find the **rank** of a score of `78` (i.e., what position is it when sorted?)
4. Print how many students scored **above the average**

```python
scores = [72, 88, 54, 95, 67, 78, 91, 63, 85, 70]
```

````python
scores = [72, 88, 54, 95, 67, 78, 91, 63, 85, 70]
````

````python
# 1. Sort descending
desc = sorted(scores, reverse=True)
print("Sorted (desc):", desc)

# 2. Top 3
print("Top 3:        ", desc[:3])

# 3. Rank of 78 (1-based)
rank = desc.index(78) + 1
print(f"Rank of 78:    {rank} out of {len(scores)}")

# 4. Above average
avg = sum(scores) / len(scores)
above_avg = [s for s in scores if s > avg]
print(f"Average:       {avg:.1f}")
print(f"Above average: {len(above_avg)} students → {above_avg}")
````

### 🏋️ Exercise — Tuples

You have GPS coordinates for three cities stored as tuples:
```python
istanbul  = (41.0082, 28.9784)
ankara    = (39.9334, 32.8597)
izmir     = (38.4192, 27.1287)
```

1. Unpack each tuple into `lat` and `lon` variables and print them nicely:
   `Istanbul → lat: 41.0082, lon: 28.9784`
2. Store all three as a **list of tuples** called `cities` with a name added:
   `("Istanbul", 41.0082, 28.9784)`
3. Sort `cities` by latitude (highest first) and print them

````python
istanbul  = (41.0082, 28.9784)
ankara    = (39.9334, 32.8597)
izmir     = (38.4192, 27.1287)

for name, coords in [("Istanbul", istanbul),("Ankara",ankara),("Izmir", izmir)]:
    lat, lon = coords
    print(f"{name} -> lat: {lat}, lon: {lon}")
````

````python
cities = [
    ("istanbul",41.0082, 28.9784),
    ("ankara" ,39.9334, 32.8597),
    ("izmir",38.4192, 27.1287)]
cities
````

````python
sorted_cities = sorted(cities, key=lambda c:c[1], reverse=True) 

for city, lat, lon in sorted_cities:
    print(f"  {city}  lat={lat}")
````

````python
either = mon | wed
print(f"Attended at least one: {either}")
````

````python
exactly_one = mon ^ wed
print(f"Exactly one: {exactly_one}")
````


### 🏋️ Exercise - Dictionaries

You're given a course grade book as a dictionary:

```python
gradebook = {
    "Alice":  [85, 92, 78, 95],
    "Bob":    [60, 55, 70, 65],
    "Carol":  [90, 95, 88, 97],
    "Dave":   [50, 60, 45, 55],
    "Eve":    [75, 80, 72, 85],
}
```

Each student has a list of 4 assignment scores. Write code to:

1. Print each student's **average score** (rounded to 1 decimal)
2. Build a new dict `averages` mapping `name → average`
3. Print the **top student** (highest average)
4. Build a dict `status` mapping `name → "Pass"` or `"Fail"` (pass if avg ≥ 65)

````python
gradebook = {
    "Alice":  [85, 92, 78, 95],
    "Bob":    [60, 55, 70, 65],
    "Carol":  [90, 95, 88, 97],
    "Dave":   [50, 60, 45, 55],
    "Eve":    [75, 80, 72, 85],
}

# part 1
for name,scores in gradebook.items():
    avg = sum(scores)/len(scores)
    print(f"Average score of {name} is {avg}")
````

````python
# part 2
averages = {name: round(sum(scores)/len(scores),1) for name,scores in gradebook.items()}
averages
````

````python
# part 3
top_student = max(averages, key=averages.get)
top_student
````

````python
# part 4
status = {name: "Pass" if avg >= 65 else "Fail" for name, avg in averages.items()}
status
````
