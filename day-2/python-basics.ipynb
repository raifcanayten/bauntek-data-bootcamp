### 🏋️ Exercise - List Comprehensions

Use list comprehensions (not regular loops) to solve the following:

**Part A:** From `numbers = [1, 2, 3, ..., 20]`, create a list of all numbers divisible by 3.

**Part B:** Given the temperatures in Celsius, convert them all to Fahrenheit.  
Formula: `F = C * 9/5 + 32`
```python
celsius = [0, 10, 20, 25, 30, 37, 100]
```

**Part C:** From the list of words below, keep only words longer than 4 characters, and return them capitalised.
```python
words = ["hi", "numpy", "for", "pandas", "data", "matplotlib", "ai", "plot"]
```

````python
# Part A — divisible by 3
div_by_3 = [x for x in range(1, 21) if x % 3 == 0]
print("Divisible by 3:", div_by_3)

# Part B — Celsius to Fahrenheit
celsius = [0, 10, 20, 25, 30, 37, 100]
fahrenheit = [c * 9/5 + 32 for c in celsius]
print("Fahrenheit:", fahrenheit)

# Part C — words longer than 4 chars, capitalised
words = ["hi", "numpy", "for", "pandas", "data", "matplotlib", "ai", "plot"]
long_caps = [w.capitalize() for w in words if len(w) > 4]
print("Long words:", long_caps)
````

## 🏆 Challenge 

You have the following student data:

```python
students = [
    {"name": "Alice",  "score": 88},
    {"name": "Bob",    "score": 54},
    {"name": "Carol",  "score": 92},
    {"name": "Dave",   "score": 67},
    {"name": "Eve",    "score": 45},
    {"name": "Frank",  "score": 79},
]
```

Write code that:
1. Prints each student's name, score, and letter grade (A/B/C/D/F) (bütün öğrencilerin  ismini, skorunu, ve harf notunu yazdır)
3. Prints the **class average** (sınıfın ortalama notunu yazdır)
4. Prints the **highest score** and which student got it (en yüksek notu alan öğrenciyi ve notunu yazdır)
5. Using a list comprehension, creates a list of names of students who **passed** (score ≥ 60) (list comprehension kullanarak dersi geçen öğrencileri, yani 60 ve üstü not alanları, yazdır.)
6. Prints the pass rate as a percentage: `"Pass rate: 66.7%"` (dersi geçme oranını yazdır.)

````python
students = [
    {"name": "Alice",  "score": 88},
    {"name": "Bob",    "score": 54},
    {"name": "Carol",  "score": 92},
    {"name": "Dave",   "score": 67},
    {"name": "Eve",    "score": 45},
    {"name": "Frank",  "score": 79},
]

# --- 1. Print each student with grade ---
print("Student Report")
print("-" * 35)
for s in students:
    score = s["score"]
    if score >= 90:   grade = "A"
    elif score >= 80: grade = "B"
    elif score >= 70: grade = "C"
    elif score >= 60: grade = "D"
    else:             grade = "F"
    print(f"  {s['name']:<8} {score:>3}  →  {grade}")

# --- 2. Class average ---
scores = [s["score"] for s in students]
average = sum(scores) / len(scores)
print(f"\nClass average: {average:.1f}")

# --- 3. Highest score ---
best = max(students, key=lambda s: s["score"])
print(f"Highest score: {best['name']} with {best['score']}")

# --- 4. List comprehension — passing students ---
passed = [s["name"] for s in students if s["score"] >= 60]
print(f"\nStudents who passed: {passed}")

# --- 5. Pass rate ---
pass_rate = len(passed) / len(students) * 100
print(f"Pass rate: {pass_rate:.1f}%")
````
