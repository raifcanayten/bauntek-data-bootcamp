```python
def calculate_grade(score, passing_mark=60):
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= passing_mark: return "D"
    else: return "F"

print(calculate_grade(85))
print(calculate_grade(55))
print(calculate_grade(55, passing_mark=50))
```

### 🏋️ Exercise — Function Basics

Write a function `bmi_report(weight_kg, height_m, name)` that:
- Computes the BMI (`weight / height²`)
- Returns a **formatted string** like:  
  `"Alice — BMI: 22.86 (Normal weight)"`
- Categories:
- `< 18.5` Underweight,
- `< 25.0` Normal weight,
- `< 30.0` Overweight,
- else Obese

Test it with: Alice (70 kg, 1.75 m), Bob (95 kg, 1.80 m), Carol (50 kg, 1.70 m)

```python
def bmi_report(weight_kg, height_m, name):
    bmi = weight_kg / (height_m ** 2)

    if bmi < 18.5:   category = "Underweight"
    elif bmi < 25.0: category = "Normal weight"
    elif bmi < 30.0: category = "Overweight"
    else:            category = "Obese"

    return f"{name} — BMI: {bmi:.2f} ({category})"


print(bmi_report(70, 1.75, "Alice"))
print(bmi_report(95, 1.80, "Bob"))
print(bmi_report(50, 1.70, "Carol"))
```

```python
# map(fn, iterable) 

# filter(fn, iterable)
```

```python
temperatures_c = [0,10,20,30,40,100]

def celcius_to_fahrenheit(c):
    return c*9/5 + 32

temps_f = list(map(celcius_to_fahrenheit,temperatures_c))
temps_f
```

```python
map(celcius_to_fahrenheit,temperatures_c)
```

```python
temps_f = list(map(lambda c: c*9/5+32,temperatures_c))
temps_f
```

```python
prices = [100, 250, 80]
quantities = [3, 1, 5 ]

totals = list(map(lambda p,q:p*q, prices, quantities))
totals
```
