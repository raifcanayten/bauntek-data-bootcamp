### 🏋️ Exercise — Array Creation

Using **only NumPy constructors** (no Python lists of numbers), create:

1. An array of integers from **1 to 20 inclusive**
2. A **3×3 identity matrix** (float64)
3. A **4×5 matrix** filled with the value `7`
4. A **2×6 matrix** of all zeros with dtype `float32`

Print each array and its `shape` and `dtype`.

```python
# 1. Integers 1–20
a = np.arange(1, 21)
print(f"1) {a}\n   shape={a.shape}, dtype={a.dtype}\n")

# 2. 3×3 identity matrix
b = np.eye(3)
print(f"2)\n{b}\n   shape={b.shape}, dtype={b.dtype}\n")

# 3. 4×5 matrix of 7s
c = np.full((4, 5), 7)
print(f"3)\n{c}\n   shape={c.shape}, dtype={c.dtype}\n")

# 4. 2×6 zeros float32
e = np.zeros((2, 6), dtype=np.float32)
print(f"5)\n{e}\n   shape={e.shape}, dtype={e.dtype}")
```

```python
# estimating pi via simulation

for N in (100, 1000, 10_000, 1_000_000):
    x = np.random.uniform(-1,1,N)
    y = np.random.uniform(-1,1,N)
    
    inside = (x**2 + y**2) <=1
    pi_estimate = 4*inside.sum() / N

    error = abs(pi_estimate -np.pi)
    print(f"Error {error} with {N} samples")
```
